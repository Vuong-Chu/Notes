```python
high_salary = (df["ConvertedComp"]>70000) & (~df["ConvertedComp"].isnull()) & (~df["LanguageWorkedWith"].isnull())
```

```python
df.loc[high_salary,["Country","Age","LanguageWorkedWith","ConvertedComp"]]
```

```python
df.loc[high_salary,["Country","Age","LanguageWorkedWith","ConvertedComp"]].dropna()
```

```python
high_salary_list = df.loc[high_salary,["Country","Age","LanguageWorkedWith","ConvertedComp"]].dropna().copy()
```

```python
high_salary_list.sort_values("ConvertedComp",ascending=False)
```

```python
countries = ["United States","United Kingdom", "India", "Germany", "Canada"]
filt = high_salary_list["Country"].isin(countries)
```

```python
high_salary_list.loc[filt].sort_values("ConvertedComp",ascending=False)
```

```python
filt = high_salary_list["LanguageWorkedWith"].str.contains("Java",na=False)
```

```python
high_salary_list.loc[filt]
```
